**Goal**: Build a deep learning model that predicts clinical toxicity of small-molecule drugs from SMILES strings, with visual explanations highlighting important substructures.

## Create Directory Structure

Let's create all necessary directories and verify the structure.



````python
import os
from pathlib import Path

# Get project root (assuming we're in notebooks/ directory)
project_root = Path().absolute().parent

# Define directories to create
directories = [
    project_root / "env",
    project_root / "notebooks",
    project_root / "src",
    project_root / "data"
]

# Create directories
for directory in directories:
    directory.mkdir(parents=True, exist_ok=True)
    print(f"✓ Created/verified: {directory}")

print("\nDirectory structure created successfully!")

````

## Verify Source Modules

Check that all source modules exist and can be imported.



````python
# Add project root to path if not already there
import sys
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

# Verify all modules exist
modules_to_check = [
    "src.data",
    "src.featurization",
    "src.models",
    "src.train",
    "src.explain",
    "src.viz",
    "src.utils"
]

print("Checking module imports...")
for module_name in modules_to_check:
    try:
        __import__(module_name)
        print(f"✓ {module_name} imported successfully")
    except ImportError as e:
        print(f"✗ {module_name} failed to import: {e}")
    except Exception as e:
        print(f"⚠ {module_name} imported but has issues: {e}")

````

## Environment Setup Instructions

### Local Setup (macOS with Conda)

1. **Create the environment:**
   ```bash
   conda env create -f env/environment.yml
   ```

2. **Activate the environment:**
   ```bash
   conda activate drug-tox-env
   ```

3. **Update the environment (if needed):**
   ```bash
   conda env update -f env/environment.yml --prune
   ```

4. **Install Jupyter kernel (for local use):**
   ```bash
   python -m ipykernel install --user --name drug-tox-env --display-name "Python (drug-tox-env)"
   ```

5. **Export requirements (optional):**
   ```bash
   pip freeze > env/requirements.txt
   ```

### Google Colab Setup

In Colab, install dependencies directly:
```python
!pip install torch-molecule torch-geometric torch-scatter deepchem transformers rdkit-pypi
```

### macOS pip3 Setup (Alternative to Conda)

If you prefer to use pip3 directly on macOS without Conda:
```bash
# Install all dependencies
pip3 install -r env/requirements.txt

# Or install individually
pip3 install torch-molecule torch-geometric torch-scatter deepchem transformers rdkit-pypi
```

**Notes**: 
- On macOS, **always use `pip3`** instead of `pip` for Python 3 packages
- The environment.yml file is primarily for local macOS setup with Conda
- Colab uses its own runtime environment and uses `pip` (not `pip3`)



````python
# Check environment.yml exists
env_file = project_root / "env" / "environment.yml"
if env_file.exists():
    print(f"✓ environment.yml found at: {env_file}")
    print("\nFirst few lines of environment.yml:")
    with open(env_file, 'r') as f:
        lines = f.readlines()[:15]
        print("".join(lines))
else:
    print(f"✗ environment.yml not found at: {env_file}")

````

## Verify Key Dependencies (if installed)

Check if essential libraries are available. Note: In a fresh environment, some may need to be installed.



````python
# Check key dependencies
dependencies = {
    "numpy": "np",
    "pandas": "pd",
    "torch": "torch",
    "rdkit": "Chem",
    "sklearn": "sklearn",
    "matplotlib": "plt",
    "seaborn": "sns"
}

print("Checking key dependencies:")
for package, alias in dependencies.items():
    try:
        exec(f"import {package} as {alias}")
        if package == "torch":
            version = torch.__version__
        elif package == "rdkit":
            from rdkit import __version__ as rdkit_version
            version = rdkit_version
        else:
            mod = eval(alias)
            version = getattr(mod, "__version__", "unknown")
        print(f"✓ {package:15s} - version: {version}")
    except ImportError:
        print(f"✗ {package:15s} - NOT INSTALLED")

# Check torch-molecule specifically
try:
    import torch_molecule
    print(f"✓ torch-molecule - installed")
except ImportError:
    print(f"✗ torch-molecule - NOT INSTALLED")
    print("  Install with: pip3 install torch-molecule (macOS) or pip install torch-molecule (Linux/Colab)")

````
